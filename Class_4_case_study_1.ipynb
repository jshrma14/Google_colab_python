```python
# Opening file for reading...

data = open("bank-data.csv")
```

```python
# Reading data from file 

data.seek(0)
lst = data.readlines()
del lst[0].  # Removed header
print(len(lst)) #Confirmed len
```

```python
# Creating set of uniq jobs

uniq_jobs = set()
for i in lst:
  print(i.split(","))
  uniq_jobs.add(i.split(",")[1])
```

```python
# Set of unique jobs
print(uniq_jobs)
```

```python
profession = input("Enter profession you want to choose..")
count_eligible = 0
if profession in uniq_jobs:
  for j in lst:
    if profession in j.split(","):
      print("This user is eligible")
      count_eligible +=1

print("Total eligible users:",count_eligible)
```
