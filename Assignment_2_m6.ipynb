```python
import numpy as np
import matplotlib.pyplot as plt
N=2000
x = np.random.uniform(-10,10,N)
y = np.random.uniform(-10,10,N)

a = np.zeros(np.size(x), dtype=bool)
b = np.sqrt(x**2+y**2)

for i in range(np.size(x)):
    if (max(abs(x[i]), abs(y[i]))>5) & (b[i] < 10):
        a[i] = True
    else:
        a[i] = False
x = x[a]
y = y[a]

plt.scatter(x,y)
plt.show()
```
